```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def make_numerical(df, cols):
    for col in cols:
        df[col] = df[col].str.rstrip("%").replace("N/A", 0)
```

```python
raw = pd.read_csv("raw-data.csv")
questions = raw.iloc[0, :]
```

```python
data = raw.iloc[2:, np.r_[9, 15:18, 19:28, 45:69]].set_index("Student Name").sort_index()
make_numerical(data, ["Q20_1", "Q20_2", "Q20_3", "Q20_4"])
data = data.astype("float")
data
```

```python
data.to_csv("Assessment_Numerical_Cleaned.csv")
```

```python
outof7 = data.iloc[:, np.r_[0:5, 12:35]]
outof100 = data.iloc[:, 5:11]
```
